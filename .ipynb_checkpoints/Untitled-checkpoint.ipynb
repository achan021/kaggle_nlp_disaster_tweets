```python
#Imports and declarations
import pandas as pd
import os
import re
import nltk
import string
import numpy as np
from numpy import array
from matplotlib import pyplot
from tqdm import tqdm
from tqdm._tqdm_notebook import tqdm_notebook
from sklearn.feature_extraction.text import CountVectorizer


#This is to enable a fancy user display for .progress_apply()
tqdm_notebook.pandas()
```

```python
#use this to check the current working directory to do relative links
#print(os.getcwd())

#increase the width of the pandas dataframe display
pd.set_option('display.width', 2000)

#Reading of the CSV file
train_csv_link = "./datasets/train.csv"
test_csv_link = "./datasets/test.csv"

train_csv = pd.read_csv(train_csv_link)
test_csv = pd.read_csv(test_csv_link)

#get the column names 
train_col_index = [index for index in train_csv.columns]
test_col_index = [index for index in test_csv.columns]

#understand the training dataset
print("The training dataset has {} rows and {} columns".format(len(train_csv) , len(train_csv.columns)))
for index in train_col_index:
    print("The number of missing data in {} : {}".format(index,train_csv[index].isnull().sum()))
```

```python
#cleaning part of the pipeline (preprocessing)
#we will clean the text labels

#define a function to parse the text into
def remove_punctuation(text):
    clean_text = "".join([char for char in text if char not in string.punctuation])
    return clean_text
#tokenize clean_text
def tokenize_text (text):
    #\W+ refers to the regex for non charcter words [^a-zA-Z0-9_]
    tokens = re.split("\W+",text)
    return tokens
#remove stopwords
def rm_stopwords(text_list):
    stopwords = nltk.corpus.stopwords.words("english")
    clean_list = [word for word in text_list if word not in stopwords]
    return clean_list
#stemming of text_token
def stem_text(text_list):
    #we use porter stemmer for now
    ps = nltk.PorterStemmer()
    stem_list = [ps.stem(word) for word in text_list]
    return stem_list

#produce clean text without punctuations
train_csv["clean_text"] = train_csv["text"].progress_apply(lambda x : remove_punctuation(x))
train_csv["text_token"] = train_csv["clean_text"].progress_apply(lambda x : tokenize_text(x))
train_csv["text_token"] = train_csv["text_token"].progress_apply(lambda x : rm_stopwords(x))
train_csv["text_token"] = train_csv["text_token"].progress_apply(lambda x : stem_text(x))



train_csv.head()
```

feature engineering ideas :
1) number of punctuations [X]
2) number of capital letters [X]
3) presence of links [X]
4) number of words [X]
5) number of hashtags [X]
6) number of weather words / natural disaster / life threat words
7) number of country words
8) number of mispelled words

```python
#count number of punctuations
def count_punc (text):
    count = 0
    for char in text:
        if char in string.punctuation:
            count += 1
    return count

#count the number of capital letters
def count_cap (text):
    count = 0
    for char in tqdm(text):
        if char.isupper() is True:
            count += 1
    return count

#boolean of links
def link_bool(text):
    if re.search("http[s]?\:\/\/",text) not in [None,"None"]:
        return 1
    return 0

#number of words
def count_words(text):
    words = len(text.split(" "))
    return words

#hashtags
def count_hash(text):
    hash_no = len(re.findall("#",text))
    return hash_no


train_csv["punc_count"] = train_csv["text"].progress_apply(lambda x : count_punc(x))
train_csv["capital_count"] = train_csv["text"].progress_apply(lambda x : count_cap(x))
train_csv["link_count"] = train_csv["text"].progress_apply(lambda x : link_bool(x))
train_csv["word_count"] = train_csv["text"].progress_apply(lambda x : count_words(x))
train_csv["hash_count"] = train_csv["text"].progress_apply(lambda x : count_hash(x))
train_csv


```

```python
#We want to check the feature importance for the labeling

print(train_csv.columns[7:])

bins = np.linspace(0,75,40)

for headers in train_csv.columns[7:-1]:
    pyplot.hist(train_csv[train_csv['target']==1][headers], bins, alpha=0.5,normed=True,label='T')
    pyplot.hist(train_csv[train_csv['target']==0][headers], bins, alpha=0.5,normed=True,label='F')
    pyplot.title(headers)
    pyplot.legend(loc='upper left')
    pyplot.show()


#from the graphs, we can see that there is slight relationship for punc_count and capital_count. the rest shows no sign of relationship...
#Therefore, we are dropping some of the cols
train_csv.drop(columns=["id","location","keyword","text","link_count", "word_count", "hash_count"],inplace=True)
train_csv
```

```python
#There are a few vectorization techniques: count [X], ngrams, TF-IDF
#lets use the count vectorizer here

def dummy(doc):
    return doc

count_vect = CountVectorizer(tokenizer = dummy, lowercase=False)
vector = count_vect.fit_transform(train_csv["text_token"])
print(vector.shape)
print(count_vect.get_feature_names())
vector_df = pd.DataFrame(vector.toarray())
vector_df
```
